```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

Importing Dataset

```python
data = pd.read_csv('datasets.csv')
```

```python
data.head(5)
```

```python
data.describe()
```

```python
data.info()
```

```python
data.isnull().sum()

#Dropping all missing rows
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()

data[data.duplicated()]
```

```python
data.drop_duplicates(inplace=True)
```

Changing data type


```python
data['id'] = data['id'].astype(object)

data.dtypes
```

```python
data['host_id'] = data['host_id'].astype(object)
```

```python
data.dtypes
```

Price Distribution

```python
df = data[data['price'] < 1500]

sns.boxenplot(data=df, x='price')
```

```python
plt.figure(figsize=(5,4))
sns.histplot(data=df, x='price')
plt.ylabel("Frequency")
```

```python
plt.figure(figsize=(6,4))
sns.histplot(data=df, x='availability_365', bins=50)
plt.title("availability_365 Distribution")
plt.ylabel('Frequency')
plt.show()
```

Feature Engineering 

```python
df['price per bed'] = df['price']/df['beds']
df.head()
```

```python
df.groupby(by='neighbourhood_group')['price'].mean()
```

```python
sns.barplot(data=df, x='neighbourhood_group', y='price',hue='room_type')
```

```python
df.groupby(by='neighbourhood_group')['price per bed'].mean()
```

```python
sns.barplot(data=df, x='neighbourhood_group', y='price per bed', hue='room_type')
```

Number of reviews and price relationship 

```python
plt.Figure(figsize=(8,5))
sns.scatterplot(data=df, x='number_of_reviews', y='price', hue='neighbourhood_group')
plt.title("Price and Reviews based on neighbourhood")
plt.show()
```

```python
df.columns
```

```python
sns.pairplot(data=df, vars=['price', 'minimum_nights', 'number_of_reviews', 'availability_365'], hue='room_type')
```

Geographical Distribution of AirBnb Listing

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='longitude', y='latitude', hue='room_type')
```
